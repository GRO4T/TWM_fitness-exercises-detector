## Detect joints using Mediapipe

```python
def detect_joints(cap, pose, i):
    ret, frame = cap.read()

    # Recolor image to RGB
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False
      
    # Make detection
    results = pose.process(image)
    
    # Recolor back to BGR
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
    # Render detections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                            mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                            mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2) 
                                )               

    image = cv2.putText(image, f"frame: {i}", (20, 20),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (124,252,0), 2)
        
    cv2_imshow(image)
```

# Data preprocessing

```python
import pickle


def process_clip(clip_path: str):
    clip_keypoints = []
    
    cap = cv2.VideoCapture(clip_path) 
    
    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
        while True:
            frame_keypoints = []

            ret, frame = cap.read()

            if frame is None:
                print(f"OK: Finished processing clip {clip_path}")
                break
        
            # Recolor image to RGB
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False
            
            # Make detection
            results = pose.process(image)
            raw_keypoints = results.pose_landmarks.landmark

            for landmark in mp_pose.PoseLandmark:
                frame_keypoints.append([
                    raw_keypoints[landmark.value].x,
                    raw_keypoints[landmark.value].y,
                    raw_keypoints[landmark.value].z,
                ])
            clip_keypoints.append(frame_keypoints) 
    
    return clip_keypoints
```

## For each video clip, convert it to a list of skeleton keypoints

```python
if IN_COLAB:
    from google.colab import auth


    auth.authenticate_user()
    project_id = "twm-projekt"
    !gcloud config set project {project_id}

X = []
Y = []

for exercise_id, exercise in enumerate(EXERCISES):
    for i in range(100):

        clip_no = str(i).zfill(6)
        clip_path = os.path.join(f"./data/infinityai_fitness_basic_{exercise}_v1.0/data", f"{clip_no}.mp4")

        print(f"{exercise_id+1}/{len(EXERCISES)} Processing clip {clip_path}... ({i}/100)")

        try:
            data = process_clip(clip_path)

            # Save converted clip
            converted_path = f"./data_preprocessed/{exercise}/{os.path.basename(clip_path)}".replace("mp4", "pkl")
            print(f"Saving converted clip to {converted_path}") 
            os.makedirs(os.path.dirname(converted_path), exist_ok=True)
            with open(converted_path, "wb") as file:
                file.write(pickle.dumps(data))

            X.append(data)
            Y.append(exercise)
        except Exception as e:
            print(f"Error processing clip {clip_path}: {e}") 

print("OK: Finished")
```

# Train an exercise classifier

```python
import pickle
import random

evaluation_ids = {exercise: random.randint(0,100) for exercise in EXERCISES}
evaluation_dataset = []

X = []
Y = []

for exercise_id, exercise in enumerate(EXERCISES):
    for i in range(100):
        if evaluation_ids[exercise] == i:
            continue

        clip_no = str(i).zfill(6)
        file_name = f"./data_preprocessed/{exercise}/{clip_no}.pkl"
        try:
            with open(file_name, "rb") as file:
                data = pickle.loads(file.read())
                X.append(data)
                Y.append(exercise_id)
        except Exception as e:
            print(f"Error processing file {file_name}: {e}")
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
print("TensorFlow version:", tf.__version__)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train_flat = []
Y_train_flat = []
X_test_flat = []
Y_test_flat = []

for i in range(len(X_train)):
    for frame in X_train[i]:
        frame2 = []
        for keypoint in frame:
            frame2.append([keypoint[0], keypoint[1]])
        X_train_flat.append(frame2)
        Y_train_flat.append(Y_train[i])

for i in range(len(X_test)):
    for frame in X_test[i]:
        frame2 = []
        for keypoint in frame:
            frame2.append([keypoint[0], keypoint[1]])
        X_test_flat.append(frame2)
        Y_test_flat.append(Y_test[i])
```

```python
import numpy as np

lstm_window_size = 5

X_train_lstm = []
Y_train_lstm = []
X_test_lstm = []
Y_test_lstm = []

for i in range(len(X_train)):
    for j in range(len(X_train[i])-5):
        window = []
        for k in range(lstm_window_size):
            frame_flat = []
            for keypoint in X_train[i][j+k]:
                frame_flat.append(keypoint[0])
                frame_flat.append(keypoint[1])
                frame_flat.append(keypoint[2])
            assert len(frame_flat) == 99
            window.append(frame_flat)
        assert len(window) == 5
        X_train_lstm.append(window)
        Y_train_lstm.append(Y_train[i])

for i in range(len(X_test)):
    for j in range(len(X_test[i])-5):
        window = []
        for k in range(lstm_window_size):
            frame_flat = []
            for keypoint in X_test[i][j+k]:
                frame_flat.append(keypoint[0])
                frame_flat.append(keypoint[1])
                frame_flat.append(keypoint[2])
            window.append(frame_flat)
        X_test_lstm.append(window)
        Y_test_lstm.append(Y_test[i])

X_train_lstm = np.array(X_train_lstm)
Y_train_lstm = np.array(Y_train_lstm)
X_test_lstm = np.array(X_test_lstm)
Y_test_lstm = np.array(Y_test_lstm)

print(X_train_lstm.shape)
print(X_test_lstm.shape)
```

```python
# 1st architecture
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(33,2)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10)
])

# 2nd architecture
# model = tf.keras.models.Sequential([
#   tf.keras.layers.LSTM(8, input_shape=(5,99)),
#   tf.keras.layers.Dense(128, activation='relu'),
#   tf.keras.layers.Dropout(0.2),
#   tf.keras.layers.Dense(2)
# ])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
# 1st architecture
model.fit(X_train_flat, Y_train_flat, epochs=5)

# 2nd architecture
#model.fit(X_train_lstm, Y_train_lstm, epochs=5)
```

# Evaluate the model

```python
score = model.evaluate(X_test_flat, Y_test_flat, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
import numpy as np
import cv2
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

text_color = (124,252,0)

#cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture("./data/infinityai_fitness_basic_superman_v1.0/data/000059.mp4") 

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while True:
        try:
            frame_keypoints = []

            ret, frame = cap.read()

            # Recolor image to RGB
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False
            
            # Make detection
            results = pose.process(image)
            raw_keypoints = results.pose_landmarks.landmark

            for landmark in mp_pose.PoseLandmark:
                frame_keypoints.append([
                    raw_keypoints[landmark.value].x,
                    raw_keypoints[landmark.value].y,
                    raw_keypoints[landmark.value].z,
                ])

            prediction = model.predict([frame_keypoints])

            number_to_class = {
                0: "armraise",
                1: "bicyclecrunch",
                2: "birddog",
                3: "curl",
                4: "fly",
                5: "legraise",
                6: "overheadpress",
                7: "pushup",
                8: "squat",
                9: "superman"
            }

            predicted_class = number_to_class[np.argmax(prediction)]

            debug_info = f"{predicted_class}: {prediction}"

            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
            image = cv2.putText(image, debug_info, (20, 20),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.4, text_color, 1)
            
            cv2.imshow("TWM", image)
        except Exception as e:
            print(f"Error processing frame: {e}")

        if cv2.waitKey(10) & 0xFF == ord('q'):
                break

cv2.destroyAllWindows()
```
